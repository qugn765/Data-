# Pandas - Series

```python
import numpy as np
import pandas as pd
```

## 시리즈 생성

```python
s = pd.Series([9904312,3448737,2890451,2466052])
s
```

```python
s = pd.Series([9904312,3448737,2890451,2466052], index=['서울','부산','인천','대구'], dtype=np.int32)
s
```

```python
# 시리즈의 
```

```python
# 시리즈의 인덱스
s.index
```

```python
s.name = '인구'
s.index.name ='도시'
s
```

## 시리즈의 연산

```python
s / 1000000
```

## 시리즈 인덱싱

```python
s[1],s['부산']
```

```python
# fancy indexing
s[[0,3,1]]
```

```python
# slicing
s[1:3]
```

```python
# 문자인덱스로 슬라이싱 하면 마지막(대구)도 포함
s['부산':'대구']
```

```python
# 필터링
s[2500000<s<3000000]
```

```python
# 필터링
s[(2500000<s)&(s<3000000)]
```

```python
s.부산
```

## 시리즈와 딕셔너리

```python
'서울' in s, '대전' in s
```

```python
s2 = pd.Series({"서울": 9631482, "부산": 3393191, "인천": 2632035, "대전": 1490158})
s2
```

## 인덱스 기반 연산

```python
s - s2
```

```python
s.values - s2.values
```

```python
# 인구 증가율
rs = (s-s2)/s2 * 100
rs
```

```python
rs.notnull()
```

```python
rs = rs[rs.notnull()]
rs
```

## 데이터의 갱신.추가.삭제

```python
# 데이터 갱신
rs['부산'] = 1.63
rs
```

```python
# 데이터 추가
rs['대구'] = 1.41
rs
```

```python
# 데이터 삭제
del rs['부산']
rs
```
