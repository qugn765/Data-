## 연습문제

```python
data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
# 1) 수학 점수를 시리즈로
df['수학']
```

```python
# 2) 모든 학생의 국어와 영어 점수를 데이터 프레임으로 나타낸다.
df[['국어','영어']]
```

```python
# 3) 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.
df.values
```

```python
np.mean(df.values, axis=1)
```

```python
df['평균'] = np.mean(df.values, axis=1).round(1)
df
```

```python
df.mean(axis=0)       #  DataFrame 함수
```

```python
# 4) 방자의 영어 점수를 80점으로 수정하고 평균 점수도 다시 계산한다.
del df['평균']
df.영어[3] = 80
df['평균'] = np.round(df.mean(axis=1),1)
df
```

```python
# 5) 춘향의 점수를 데이터프레임으로 나타낸다.
df[:'춘향']
```

```python
df.T[['춘향']]
```

```python
# 중간에 있는 두 행 꺼내오기
df.head(3).tail(2)
```

```python
# 6) 향단의 점수를 시리즈로 나타낸다.
df.T['향단']
```
